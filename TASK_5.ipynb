```python
#1.How many people have heart diseases and how many people doesn't have heart diseases?
df.target.value_counts()
```

```python
df.target.value_counts().plot(kind = 'bar',color=["orchid","salmon"])
plt.title("heart disease values")
plt.xlabel("1=Heart Disease,0 = no heart disease")
plt.ylabel("amount");
```

```python
df.target.value_counts().plot(kind='pie',figsize=(5,5))
plt.legend(["Disease","No Disease"]);
```

```python
#now lets check how many 'male' and 'female'are in the dataset
df.sex.value_counts()
```

```python
df.sex.value_counts().plot(kind = 'pie',figsize=(5,5))
plt.title('Male Female ratio')
plt.legend(['Male','Female']);
```

```python
#2.people of which sex has most heart disease?
pd.crosstab(df.target,df.sex)
```

```python
sns.countplot(x='target',data=df,hue='sex')
plt.title("heart disease frequency for sex")
plt.xlabel("0=No heart disease,1=heart disease");
```

```python
#number of male is more tham double in our dataset than female.
#more than '45%'male has heart disease and '75% female'has heartdisease.
```

```python
#3.people of which sex has which type of chest pain most?
df.cp.value_counts()
```

```python
df.cp.value_counts().plot(kind='bar',color=['salmon','lightskyblue','green','khaki'])
plt.title('chest pain type vs count');
```

```python
pd.crosstab(df.sex,df.cp)
```

```python
pd.crosstab(df.sex,df.cp).plot(kind = 'bar',color=['coral','blue','springgreen','plum'])
plt.title('Type of chest pain for sex')
plt.xlabel('0=Female, 1=male')
```

```python
#most of male has type 0 chest pain and least of male has type 4 pain
#in case of 'female' type 0 and type 2 percentage is alomost same.
```
